```python
import numpy as np
```

# axis

```python
x = np.array([[1, 2], [3, 4]], dtype=np.float64)
y = np.array([[5, 6], [7, 8]], dtype=np.float64)
```

```python
print(x + y)
print(np.add(x , y))
```

```python
print(x - y)
print(np.subtract(x, y))
```

```python
print(x * y)
print(np.multiply(x, y))
```

```python
print(x / y)
print(np.divide(x, y))
```

```python
print(x @ y)
print(np.matmul(x, y))
```

```python
z = np.arange(1, 11).reshape(2, 5)
print(z)
```

axis는 각 배열의 차원에 해당하는 index.  
axis=0으로 sum을 연산하는 것은 0번 축이 사라지는 방향으로 원소들을 모두 더하라는 의미.

```python
sum0 = np.sum(z)
sum1 = np.sum(z, axis=0)
sum2 = np.sum(z, axis=1)
sum3 = np.sum(z, axis=-1)

print(sum0, sum0.shape)
print(sum1, sum1.shape)
print(sum2, sum2.shape)
print(sum3, sum3.shape)
```

```python
print(z[0, :] + z[1, :]) ## axis=0이면, (2, 5) shape에서 0번째 2가 없어지는 방향으로 연산.
```

# Broadcasting


```python
x = np.array([1, 2, 3]).reshape(1, 3)
y = np.array([4, 5]).reshape(1, 2)

print(x, x.shape)
print(y, y.shape) ## broadcast가 불가능함.
```

```python
y = y.swapaxes(0, 1)
print(y, y.shape)
```

```python
print(x + y)
```

# reshape

```python
a = np.arange(24).reshape(2, 3, 4)
print(a, a.shape)
```

```python
a = a.reshape(4, -1) ## 4,6
print(a, a.shape)
```

```python
# a = a[:, :, new_axis]
a = a[..., np.newaxis] ## 차원을 추가한다.
print(a, a.shape)
```

```python
a = a.reshape(4, 6)
a = np.expand_dims(a, axis=0)
print(a, a.shape)
```

```python
b = a.copy()
print(b, b.shape)
```

```python
c = np.concatenate((a, b), axis=-1) ## concatenate은 지정된 축을 기준으로 배열을 이어 붙임
print(c, c.shape)
```

```python
c = np.concatenate((a, b), axis=0)
print(c, c.shape)
```

```python
d = np.stack((a, b), axis=-1) ## stack은 신규 차원을 추가하면서 배열을 축적하는 개념
print(d, d.shape)
```

```python
d = np.stack((a, b), axis=0)
print(d, d.shape)
```
