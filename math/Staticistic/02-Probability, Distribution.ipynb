```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import norm
from scipy.stats import uniform
```

```python
data_dir = "/home/pervinco/DL-basic/data/weight-height.csv"
data = pd.read_csv(data_dir)

print(data)
```

```python
heights = data["Height"]
```

```python
# 막대그래프의 구간 크기 설정 (10cm 단위)
bin_size = 10
bins_range = range(int(heights.min()), int(heights.max()) + bin_size, bin_size)

# 막대그래프 그리기
plt.figure(figsize=(12, 6))
hist, bins, _ = plt.hist(heights, bins=bins_range, color='skyblue', edgecolor='black', alpha=0.7, label='height distribution')

# 각 구간의 중앙값 계산
bin_centers = 0.5 * (bins[1:] + bins[:-1])

# 중앙값을 이용해 막대그래프들을 연결하는 그래프 그리기
plt.plot(bin_centers, hist, color='red', marker='o', linestyle='-', linewidth=2, markersize=8, label='class value')

plt.xlabel('height (cm)')
plt.ylabel('class')
plt.title('Continuous Distribution Graph')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 막대그래프의 구간 크기 설정 (10cm 단위)
bin_size = 1
bins_range = range(int(heights.min()), int(heights.max()) + bin_size, bin_size)

# 막대그래프 그리기
plt.figure(figsize=(12, 6))
hist, bins, _ = plt.hist(heights, bins=bins_range, color='skyblue', edgecolor='black', alpha=0.7, label='height distribution')

# 각 구간의 중앙값 계산
bin_centers = 0.5 * (bins[1:] + bins[:-1])

# 중앙값을 이용해 막대그래프들을 연결하는 그래프 그리기
plt.plot(bin_centers, hist, color='red', marker='o', linestyle='-', linewidth=2, markersize=8, label='class value')

plt.xlabel('height (cm)')
plt.ylabel('class')
plt.title('Continuous Distribution Graph')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 표준 정규 분포에 대한 확률 밀도 함수 (PDF)를 계산할 값의 범위
x = np.linspace(-3, 3, 1000)

# 평균 0, 표준편차 1의 정규 분포 확률 밀도 함수
pdf = norm.pdf(x, 0, 1)

# 확률 밀도 함수를 그리는 그래프
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Probability (P(X = x))', linewidth=2)
plt.title('Normal Distribution, X ~ N(0, 1)')
plt.xlabel('Random Variable (X)')
plt.ylabel('Probability (P(X = x))')
plt.grid(True)
plt.show()
```

```python
# 여러 평균과 표준편차에 대한 정규 분포 그래프
plt.figure(figsize=(12, 8))

# 다양한 평균(μ)과 표준편차(σ)를 가진 정규 분포
means = [0, 0, -2, 2]  # 평균의 변화
std_devs = [0.5, 1.5, 0.5, 0.5]  # 표준편차의 변화
colors = ['blue', 'green', 'red', 'orange']
labels = [
    'μ = 0, σ = 0.5',
    'μ = 0, σ = 1.5',
    'μ = -2, σ = 0.5',
    'μ = 2, σ = 0.5'
]

for mean, std_dev, color, label in zip(means, std_devs, colors, labels):
    x = np.linspace(-5, 5, 1000)
    pdf = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    plt.plot(x, pdf, label=label, linewidth=2, color=color)

plt.title('Normal Distribution Variations')
plt.xlabel('Random Variable (X)')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 균등 분포에 대한 값의 범위
x = np.linspace(-3, 3, 1000)

# 최소값 -3, 최대값 3의 균등 분포 확률 밀도 함수
pdf = uniform.pdf(x, -3, 6) # uniform.pdf(x, loc, scale)에서 loc는 최소값, scale은 범위의 길이

# 확률 밀도 함수를 그리는 그래프
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Probability (P(X = x))', linewidth=2, color='red')
plt.title('Uniform Distribution, X ~ U(-3, 3)')
plt.xlabel('Random Variable (X)')
plt.ylabel('Probability (P(X = x))')
plt.grid(True)
plt.show()
```

```python
# 정규 분포 설정
mu, sigma = 0, 0.1 # 평균과 표준편차
normal_dist = np.random.normal(mu, sigma, 1000)

# 정규 분포 그래프
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1) # 1행 2열의 첫 번째
plt.hist(normal_dist, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# 균등 분포 설정
a, b = -1, 1 # 최소값과 최대값
uniform_dist = np.random.uniform(a, b, 1000)

plt.figure(figsize=(10, 5))

# 균등 분포 그래프
plt.subplot(1, 2, 2) # 1행 2열의 두 번째
plt.hist(uniform_dist, bins=30, density=True, alpha=0.6, color='r')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
