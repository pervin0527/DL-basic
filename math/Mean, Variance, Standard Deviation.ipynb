```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dir = "/home/pervinco/DL-basic/data/2017_seoul_temperature-2C.csv"
df = pd.read_csv(data_dir)

print(df)
```

```python
days = df["Days"]
temp = df["AvgTemperature"]

print(days.shape, temp.shape)
```

```python
plt.plot(days, temp, 'ro', markersize=3, label="original")
plt.xlabel("days")
plt.ylabel("temperature")
plt.legend()
plt.show()
```

<a id="1"></a>
## 1.Mean

$$ \bar x = \frac{1}{N}\sum_{i=1}^N x_i $$

```python
temp_avg = np.sum(temp) / len(temp)
print(temp_avg)
```

```python
temp_avg = np.mean(temp, axis=0)
print(temp_avg)
```

```python
avgs = np.repeat(temp_avg, len(days))
print(avgs.shape)
print(avgs[:10])

plt.plot(days, temp, 'ro', markersize=3, label="original")
plt.plot(days, avgs, color="#FF6B33", markersize=3, label="mean")

plt.xlabel("days")
plt.ylabel("temperature")
plt.legend()
plt.show()
```

<a id="2"></a>
## 2. Variance

$$ \text{var(X)}=\frac{1}{N}\sum_{i=1}^N(x_i-\bar x)^2 $$

```python
temp_avg = np.mean(temp, axis=0)
temp_var = np.mean((temp - temp_avg) ** 2)
print(temp_var)
```

```python
temp_var = np.var(temp, axis=0)
print(temp_var)
```

```python
vars = np.repeat(temp_var, repeats=len(days))

plt.plot(days, temp, 'ro', markersize=3, label="original")
plt.plot(days, vars, color="#FF6B33", markersize=3, label="variance")

plt.xlabel("days")
plt.ylabel("temperature")
plt.legend()
plt.show()
```

<a id="3"></a>
## 3.Standard Deviation

```python
temp_avg = np.sum(temp) / len(temp)
temp_var = np.sum((temp - temp_avg) ** 2) / len(temp)
temp_std = np.sqrt(temp_var)

print(temp_std)
```

```python
temp_std = np.std(temp, axis=0)
print(temp_std)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(days, temp, 'ro', markersize=3, label="original")
plt.plot(days, avgs, color="#FF6B33", markersize=3, label="mean")
plt.fill_between(days, temp_avg, temp, color='grey', alpha=0.2, label='Deviation from Mean')

plt.title('Daily Average Temperature')
plt.xlabel('Day of the Year')
plt.ylabel('Average Temperature (°C)')
plt.legend()
plt.show()
```
