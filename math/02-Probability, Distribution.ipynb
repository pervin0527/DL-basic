```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import norm
from scipy.stats import uniform
```

```python
data_dir = "/home/pervinco/DL-basic/data/weight-height.csv"
data = pd.read_csv(data_dir)

print(data)
```

```python
heights = data["Height"]
```

```python
#     (10cm )
bin_size = 10
bins_range = range(int(heights.min()), int(heights.max()) + bin_size, bin_size)

#  
plt.figure(figsize=(12, 6))
hist, bins, _ = plt.hist(heights, bins=bins_range, color='skyblue', edgecolor='black', alpha=0.7, label='height distribution')

#    
bin_centers = 0.5 * (bins[1:] + bins[:-1])

#      
plt.plot(bin_centers, hist, color='red', marker='o', linestyle='-', linewidth=2, markersize=8, label='class value')

plt.xlabel('height (cm)')
plt.ylabel('class')
plt.title('Continuous Distribution Graph')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#     (10cm )
bin_size = 1
bins_range = range(int(heights.min()), int(heights.max()) + bin_size, bin_size)

#  
plt.figure(figsize=(12, 6))
hist, bins, _ = plt.hist(heights, bins=bins_range, color='skyblue', edgecolor='black', alpha=0.7, label='height distribution')

#    
bin_centers = 0.5 * (bins[1:] + bins[:-1])

#      
plt.plot(bin_centers, hist, color='red', marker='o', linestyle='-', linewidth=2, markersize=8, label='class value')

plt.xlabel('height (cm)')
plt.ylabel('class')
plt.title('Continuous Distribution Graph')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#        (PDF)   
x = np.linspace(-3, 3, 1000)

#  0,  1     
pdf = norm.pdf(x, 0, 1)

#     
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Probability (P(X = x))', linewidth=2)
plt.title('Normal Distribution, X ~ N(0, 1)')
plt.xlabel('Random Variable (X)')
plt.ylabel('Probability (P(X = x))')
plt.grid(True)
plt.show()
```

```python
#       
plt.figure(figsize=(12, 8))

#  () ()   
means = [0, 0, -2, 2]  #  
std_devs = [0.5, 1.5, 0.5, 0.5]  #  
colors = ['blue', 'green', 'red', 'orange']
labels = [
    ' = 0,  = 0.5',
    ' = 0,  = 1.5',
    ' = -2,  = 0.5',
    ' = 2,  = 0.5'
]

for mean, std_dev, color, label in zip(means, std_devs, colors, labels):
    x = np.linspace(-5, 5, 1000)
    pdf = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    plt.plot(x, pdf, label=label, linewidth=2, color=color)

plt.title('Normal Distribution Variations')
plt.xlabel('Random Variable (X)')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#     
x = np.linspace(-3, 3, 1000)

#  -3,  3     
pdf = uniform.pdf(x, -3, 6) # uniform.pdf(x, loc, scale) loc , scale  

#     
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Probability (P(X = x))', linewidth=2, color='red')
plt.title('Uniform Distribution, X ~ U(-3, 3)')
plt.xlabel('Random Variable (X)')
plt.ylabel('Probability (P(X = x))')
plt.grid(True)
plt.show()
```

```python
#   
mu, sigma = 0, 0.1 #  
normal_dist = np.random.normal(mu, sigma, 1000)

#   
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1) # 1 2  
plt.hist(normal_dist, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#   
a, b = -1, 1 #  
uniform_dist = np.random.uniform(a, b, 1000)

plt.figure(figsize=(10, 5))

#   
plt.subplot(1, 2, 2) # 1 2  
plt.hist(uniform_dist, bins=30, density=True, alpha=0.6, color='r')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
